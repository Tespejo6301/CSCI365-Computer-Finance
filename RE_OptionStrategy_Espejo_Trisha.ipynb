### Option Trading Strategy

[**Options Strategy Guide from Fidelity**](https://www.fidelity.com/learning-center/investment-products/options/options-strategy-guide/overview)

**Define payoff for building blocks**

```python
def long_call(S, K, premium):
    return (max(0, S-K) - premium)
def short_call(S, K, premium):
    return (premium - max(0, S - K))
def long_put(S, K, premium):
    return ( max(0, K - S) - premium)
def short_put(S, K, premium):
    return ( premium - max(0, K-S))
```

**Question 1. Plot Payoff diagram for a long Strangle using the following link as example. What are the breakeven price?**

[Long Strangle](https://www.fidelity.com/learning-center/investment-products/options/options-strategy-guide/long-strangle)

```python
K1 = 105
premium1 = 1.50
K2 = 95 
premium2 = 1.30
X = range(90, 111)
P = [ long_call(S,K1, premium1) +  long_put(S,K2, premium2) for S in X]
```

```python
sns.set_style("whitegrid", {"grid.linestyle": '--'})
sns.lineplot(x=X, y=P)
```

**Calculating the breakeven price**

```python
## net cost 
net_cost = premium1 + premium2

## high break even price 
high_breakeven_price = K1 + net_cost

## low break even price 
low_breakeven_price = K2 - net_cost


print("net_cost:",net_cost)
print("high breakeven price:", high_breakeven_price)
print("low breakeven price: ", low_breakeven_price)
```

There are two breakeven prices the high and low. For the high it is calcualted by K1 + net cost (K1 is the value of the long call). While for the low it is calculated by K2 - net cost (k2 is value of the long put). For the high breakeven price I got 107.8 and for the low breakeven I got 92.2.

**Question 2. Plot Payoff diagram for a short iron condor pread with calls using the following link as example. What are the breakeven price?**

[Short Iron Condor Spread](https://www.fidelity.com/learning-center/investment-products/options/options-strategy-guide/short-iron-condor-spread)

```python
K1 = 95
premium1 = 0.70
X = range(90, 116)
K2 = 100
premium2 = 2.10
K3 = 105
premium3 = 2.35
K4 = 110
premium4 = 0.95
P = [ long_put(S, K1, premium1) + short_put(S, K2, premium2) + short_call(S,K3, premium3)+ long_call(S,K4, premium4)   for S in X]
```

```python
sns.set_style("whitegrid", {"grid.linestyle": '--'})
sns.lineplot(x=X, y=P)
```

**Calculating the breakeven price**

```python
## net cost 
net_cost = 2.8

## high break even price 
high_breakeven_price = K3 + net_cost

## low break even price 
low_breakeven_price = K2 - net_cost


print("net_cost:",net_cost)
print("high breakeven price:", high_breakeven_price)
print("low breakeven price: ", low_breakeven_price)
```

 The lower breakeven point is the stock price equal to the strike price of the short put minus the net credit received. The upper breakeven point is the stock price equal to the strike price of the short call plus the net credit received. For the high breakeven price I got 107.8 and for the low breakeven I got 97.2.
